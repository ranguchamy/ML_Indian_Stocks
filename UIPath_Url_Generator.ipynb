```python
# Base URL
base_url = "https://trendlyne.com/get-document/IndiaDailyDeals/file/"

# Generate 100 incremented URLs
incremented_urls = [base_url + str(i) + "/" for i in range(12892, 12992)]

# Sort the URLs by the highest number
sorted_urls = sorted(incremented_urls, key=lambda x: int(x.split("/")[-2]), reverse=True)

# Display the sorted URLs
for url in sorted_urls:
    print(url)
```
