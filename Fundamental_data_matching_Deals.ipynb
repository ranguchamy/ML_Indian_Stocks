```python
import pandas as pd
from tabulate import tabulate

def get_stock_info(nse_code, file_path):
    # Read the Excel file into a DataFrame
    df = pd.read_excel(file_path)
    # print(df)
    # Filter the DataFrame based on the NSE code
    filtered_df = df[df['NSE code'] == nse_code]

    if filtered_df.empty:
        return f"No data found for NSE code {nse_code}"

    # Select only the required columns
    result_df = filtered_df[['Stock Name','Industry Name', 'Current Price Rs', 'Price to Book Value', 'Industry Price to Book TTM','PE TTM Price to Earnings','RSI','Long Term Debt To Equity Annual','ROE Annual %','% Distance from 52week high','EPS Annual Rs','FII holding current Qtr %']]

    return result_df

# Example usage:
nse_code = 'ITC'
file_path = r'C:\Users\91908\Documents\Raja\Share market\Analysis\Trendlyne\Data\Fundamental\NSE_Fundamental.xlsx'

result = get_stock_info(nse_code, file_path)
print(tabulate(result, headers='keys', tablefmt='pretty', showindex=False))
```

```python
import pandas as pd
from tabulate import tabulate
import seaborn as sns

def get_stock_info(nse_code, file_path):
    # Read the Excel file into a DataFrame
    df = pd.read_excel(file_path)
    df_correlation = df[["PE TTM Price to Earnings","EPS Annual Rs","ROE Annual %"]]
    print(df_correlation.corr())

    sns.heatmap(data=df_correlation.corr(),
                annot=True,
                fmt='.1g',
                center=0.002536,
                cmap='coolwarm')

# Example usage:
nse_code = 'ITC'
file_path = r'C:\Users\91908\Documents\Raja\Share market\Analysis\Trendlyne\Data\Fundamental\NSE_Fundamental.xlsx'

result = get_stock_info(nse_code, file_path)
print(tabulate(result, headers='keys', tablefmt='pretty', showindex=False))
```
