```python
import os
import shutil
import re

def move_files(source_dir, destination_dir):
    # Ensure the destination directory exists
    os.makedirs(destination_dir, exist_ok=True)

    # Define the regex pattern to match the desired file name format
    pattern = re.compile(r'\d{8}-DailyDeals-Trendlyne\.xlsx')

    # Iterate through files in the source directory
    for file_name in os.listdir(source_dir):
        # Check if the file name matches the pattern
        if pattern.match(file_name):
            # Construct the full paths
            source_path = os.path.join(source_dir, file_name)
            destination_path = os.path.join(destination_dir, file_name)

            # Move the file
            shutil.move(source_path, destination_path)
            print(f"File moved: {file_name}")

# Example usage
source_directory = r'C:\Users\91908\Documents\Raja\Share market\Analysis\Trendlyne\Data'
destination_directory = r'C:\Users\91908\Documents\Raja\Share market\Analysis\Trendlyne\Data\Inbound'

move_files(source_directory, destination_directory)
```

```python
import os
import shutil
import re

def move_files(source_dir, destination_dir, file_name_format_pattern):
    # Ensure the destination directory exists
    os.makedirs(destination_dir, exist_ok=True)

    # Get a list of files to move
    files_to_move = [file for file in os.listdir(source_dir) if file_name_format_pattern.match(file)]

    # Check if there are files to move
    if not files_to_move:
        print("No files matching the specified pattern to move.")
        return

    # Iterate through files in the source directory
    for file_name in files_to_move:
        # Construct the full paths
        source_path = os.path.join(source_dir, file_name)
        destination_path = os.path.join(destination_dir, file_name)

        # Move the file
        shutil.move(source_path, destination_path)
        print(f"File moved: {file_name} to {destination_directory}")

    print("Files moved successfully.")

# Source directory
source_directory = r'C:\Users\91908\Documents\Raja\Share market\Analysis\Trendlyne\Data'

# Destination directory
destination_directory = r'C:\Users\91908\Documents\Raja\Share market\Analysis\Trendlyne\Data\OutboundSuperstars'

# Define the file name format pattern
file_name_format_pattern = re.compile(r'\d{8}-Superstars-Trendlyne\.xlsx')

# Move files using the function
move_files(source_directory, destination_directory, file_name_format_pattern)
```
